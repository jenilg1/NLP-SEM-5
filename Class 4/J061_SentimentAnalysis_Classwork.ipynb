```
import pandas as pd
import nltk
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.metrics import confusion_matrix
from sklearn.metrics import confusion_matrix,accuracy_score
```

```
nltk.download('vader_lexicon')
```

```
data=pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz",lines=True)
```

```
data.head()
```

```
data1=pd.read_csv("https://raw.githubusercontent.com/zfz/twitter_corpus/master/full-corpus.csv")
```

```
data1.head()
```

```
def sentpol(datax):
  return TextBlob(datax).sentiment.polarity
```

```
data['Sentiment_polarity']=data['reviewText'].apply(sentpol)
```

```
data1['Sentiment_pol']=data1['TweetText'].apply(sentpol)
```

```
def sentblob(datax):
  aa=TextBlob(datax).sentiment
  pol=aa.polarity
  if pol>0:
     return "positive"
  elif pol==0:
    return "neutral"
  elif pol<0:
    return "negative"
```

```
data['TextBlob_Sentiment']=data['reviewText'].apply(sentblob)
```

```
data1['TextBlob_Sentiment']=data1['TweetText'].apply(sentblob)
```

```
sentblob(data['reviewText'][190])
```

```
sentblob(data1['TweetText'][190])
```

```
def sentnltk(datax):
  sia=SentimentIntensityAnalyzer()
  s=sia.polarity_scores(datax)
  if s['compound'] <= -0.05:
    return "negative"
  elif s['compound'] >= 0.05:
    return "positive"
  elif -0.5<s['compound']<0.5:
    return "neutral"
```

```
data['NLTK_Sentiment']=data['reviewText'].apply(sentnltk)
```

```
data1['NLTK_Sentiment']=data1['TweetText'].apply(sentnltk)
```

```
data.head()
```

```
data1.head()
```

```
data1 = data1[data1['Sentiment'] != "irrelevant"]
data1['Sentiment'].value_counts()
```

```
blobmatrix = confusion_matrix(data1['Sentiment'], data1['TextBlob_Sentiment'])
nltkmatrix = confusion_matrix(data1['Sentiment'], data1['NLTK_Sentiment'])
```

```
blobacc = accuracy_score(data1['Sentiment'], data1['TextBlob_Sentiment'])
nltkacc = accuracy_score(data1['Sentiment'], data1['NLTK_Sentiment'])
```

```
print('TextBlob Confusion Matrix for Twitter Data is:','\n',blobmatrix,'and accuracy is:',blobacc)
```

```
print('NLTK Confusion Matrix for Twitter Data  is:','\n',nltkmatrix,'and TextBlob accuracy is:',nltkacc)
```
