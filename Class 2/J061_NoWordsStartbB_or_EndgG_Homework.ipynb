# **Class 2: No words starting with b and ending with g**

```
import pandas as pd
import re
```

```
data = pd.read_json('http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz', lines=True) 
```

```
data.head()
```

```
def nob_nog(data1):
    data1 = data1.lower() 
    final = re.findall(r'\b(?!b)\w+(?<!g)\b', data1) 
    return (final)
```

```
data['reviewText'].apply(nob_nog)
```
