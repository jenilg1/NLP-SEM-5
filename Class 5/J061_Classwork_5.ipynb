```
data2 = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz" , lines=True)
```

```
stop=stopwords.words('english')
```

```
punct = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
```

```
def punc(str):
  no_punc = ""
  for char in str:
    if char not in punct:
      no_punc = no_punc + char
  return no_punc
```
